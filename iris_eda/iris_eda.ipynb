# IRIS Dataset


```python
import pandas as pd
import seaborn as sns
```

```python
#read data det
df = pd.read_csv('iris.csv')
```

```python
df.describe()# Summary statistics
```

```python
df.isnull()
# missing/null values
```

```python
df.dtypes # Display datatype of each column
```

```python
df['sepal_area'] = df['sepal_length'] * df['sepal_width'] # Add new column 'sepal_area'
df.head()
```

```python
se = df[df['species']=='setosa']#new dataset for setosa
se
```

```python
df.sort_values(by='petal_length',ascending=False)#sortin asending order
```

```python
petal_len = df[df['petal_length']>1.5]
petal_len
#Filter data for petal width > 1.5 cm
```

```python
meean_width = df.groupby('species')['sepal_width'].mean()
meean_width
#mean sepal width for each species
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='sepal_length', bins=10, color='skyblue')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
#  histogram of the 'sepal_length' for the entire dataset
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
#scatter plot to visualize the relationship between 'sepal_length' and 'sepal_width'
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='species', y='petal_length')
plt.title('Distribution of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.show()
#boxplot to show the distribution of 'petal_length' for each species
```

```python
import matplotlib.pyplot as plt
sns.pairplot(data=df, hue='species', palette='Set2')
plt.suptitle('Pairplot of Iris Dataset')
# pairplot for all numerical columns, colored by species
```

# Observations

```python
---Versicolor and Virginica have a wider range of sepal lengths, with some specimens 
having longer sepals than any Setosa specimen.
---Versicolor generally has intermediate petal lengths, 
while Virginica tends to have the longest petal lengths among the three species.
---Setosa species tend to cluster with shorter sepal lengths and broader sepal widths compared to Versicolor and Virginica.
---Setosa species display distinct features, particularly in petal length and width, 
setting them apart from Versicolor and Virginica. Versicolor and Virginica, however,
often exhibit overlapping feature distributions, making it challenging to differentiate them
solely based on individual features.
```
